
 # Recommendation System

A Recommendation System is an AI-based tool that suggests items (like movies, products, or songs) to users based on their preferences or behavior.


üîπ Types of Recommendation Systems

1. Content-Based Filtering

Recommends items similar to what the user liked before.
Example: Netflix suggests movies with the same genre or actors you watched.

2. Collaborative Filtering

Recommends based on similar users‚Äô choices.
Example: ‚ÄúPeople who liked this also liked that.‚Äù

3. Hybrid System

Combines content-based + collaborative methods for better accuracy.
Example: YouTube uses both user behavior and video features.

now we apply pre processsing using helper function

=> initially the given object is string after using ast.literal it will convert it to the list so that help in fetching above output

```python
import ast
def convert(obj):
    
    L=[]
    for i in ast.literal_eval(obj):
       L.append( i['name'])
        
    return L    
        
```

```python
movies_credits['genres']=movies_credits['genres'].apply(convert)
```

```python
movies_credits.head()
```

```python
movies_credits['keywords']=movies_credits['keywords'].apply(convert)
```

```python
movies_credits.head()
```

```python
import ast
def convert3(obj):
    counter=0
    L=[]
    for i in ast.literal_eval(obj):
        
        if counter!=3:
            
            L.append( i['name'])
            counter+=1
        else:
            break
    return L    
        
```

```python
movies_credits['cast']=movies_credits['cast'].apply(convert3)
```

```python
movies_credits.head()
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_credits['crew']=movies_credits['crew'].apply(fetch_director)
```

```python
import ast

def fix_tags(x):
    if isinstance(x, str):
        try:
            x = ast.literal_eval(x)   # convert stringified list to real list
        except:
            pass
    if isinstance(x, list):
        return " ".join(x)
    return str(x)
```
